```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_parquet("employee_earnings.parquet")
```

```python
df
```

```python
def generate_data(a):
    min_values = df['earnings'].min()
    average_values = df['earnings'].mean()
    
    for i in range(a):
        df['earnings'] = np.random.randint(min_values, average_values, size=df.shape[0])
        filename = f"dataframe{i}.parquet"  # Use string formatting to create the filename
        df.to_parquet(filename)
        return df
```

```python
print(generate_data(2))
```
